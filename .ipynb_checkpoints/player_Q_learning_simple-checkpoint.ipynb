```python
import numpy as np
import weight_loss_player
import matplotlib.pyplot as plt
from std_policies import eps_greeedy_policy
from mpl_toolkits.mplot3d import Axes3D
from tqdm import tqdm
%matplotlib inline
```

```python
player = weight_loss_player.WeighLossPlayer()
action_space_description,obesrvation_space_description = player.get_space_description()
n_action_states = len(action_space_description)
n_obs_states =len(obesrvation_space_description)
```

```python
print('Activity Levels: {0}'.format(player.activity_levels))
print('Calorie Intake Levels: {0}'.format(player.calorie_intake_levels))
print('Calorie Quality Levels: {0}'.format(player.calorie_quality_levels))
print('Action Space: {0}'.format(n_action_states))
print('Observation Space: {0}'.format(n_obs_states))
```

```python
# Initialization for Q-Learning
Q=np.zeros((n_obs_states,n_action_states))
gamma = 0.9
lr_td = 0.8
epsilon = 0.1
n_episodes = 1000
n_steps=200
```

```python
# Initialization for recording keeping
data_log={}
episode_rewards=[]
episode_steps=[]
```

```python
# Learning
for episode in tqdm(range(n_episodes)):
    obs = player.reset()
    obs_index=obesrvation_space_description.index(obs)
    episode_log=[]
    episode_reward=0
    
#     print("Episide {0}  of {1}".format(episode,n_episodes))
    for step in range(n_steps):
        # Determine Action
        action_index =  eps_greeedy_policy(Q[obs_index,:],eps=epsilon)
        action = action_space_description[action_index]
        
        # Take ation and get reward
        obs_next,reward,exit_cond = player.step(action)
        obs_next_index = obesrvation_space_description.index(obs_next)
        #print(obs_next,reward,exit_cond)
        # log data
        hidden_states = (player.satiety_t,player.motivation_t)
        
        episode_log.append([step,obs_index,action_index,reward,obs_next_index,hidden_states])
        episode_reward += reward
        
        # Update Q
        td_target = reward + gamma*np.max(Q[obs_next_index,:])
        td_error = td_target - Q[obs_index,action_index]
        Q[obs_index,action_index] = Q[obs_index,action_index] + lr_td*(td_error)
        
        # Update States
        obs_index = obs_next_index
        
        if exit_cond:
#             print("Exiting episode based on exit conditon at step {0}.".format(step))
#             print("Final state {0}".format(obesrvation_space_description[obs_index]))
            if step==0:
                print(step,obs_index,action_index,reward,obs_next_index)
            break
#         if step == n_steps-1:
#             print("Final state {0}".format(obesrvation_space_description[obs_index]))
    data_log[episode]=episode_log
    episode_rewards.append(episode_reward)
    episode_steps.append(step)
    
        
```

```python
avg_accept_flag=[]
for episode in range(n_episodes):
    acpt_flag = [obesrvation_space_description[x[1]][1] for x in data_log[episode]]
    avg_accept_flag.append(np.sum(acpt_flag)*1./len(acpt_flag))
f,ax=plt.subplots(3,1,figsize=(15, 15))
ax[0].plot(episode_steps)
ax[1].plot(episode_rewards)
ax[2].plot(avg_accept_flag)

ax[0].set_title('Episode Steps')
ax[1].set_title('Episode Reward')
ax[2].set_title('Episode avg_accept_flag')
plt.figure()
plt.plot(episode_steps,episode_rewards,'.')
plt.xlabel('Number of episodes')
plt.ylabel('Episode Reward')
```

```python
## Plots

f,ax=plt.subplots(4,1,figsize=(15, 10))
episodes_to_plot=[0,24,40,74,99]
for plot_episode in episodes_to_plot:
    wt = [obesrvation_space_description[x[1]][0] for x in data_log[plot_episode]]
    acpt_flag = [obesrvation_space_description[x[1]][1] for x in data_log[plot_episode]]
    
    activity_level=[action_space_description[x[2]][0] for x in data_log[plot_episode]]
    cal_level=[action_space_description[x[2]][1] for x in data_log[plot_episode]]
    cal_qual_level=[action_space_description[x[2]][2] for x in data_log[plot_episode]]
    
    
    
    r = [x[3] for x in data_log[plot_episode]]
    
    satiety=[x[5][0] for x in data_log[plot_episode]]
    motivation=[x[5][1] for x in data_log[plot_episode]]

    
    ax[0].plot(wt,label='Episode {0}'.format(plot_episode))
    ax[1].plot(satiety,label='Episode {0}'.format(plot_episode))
    ax[2].plot(motivation,label='Episode {0}'.format(plot_episode))
    ax[3].plot(r,label='Episode {0}'.format(plot_episode))
    
ax[0].legend(fontsize=20)
ax[0].set_title('Weight lbs')
ax[1].set_title('satiety')
ax[2].set_title('Motivation')
ax[3].set_title('Reward')
```
