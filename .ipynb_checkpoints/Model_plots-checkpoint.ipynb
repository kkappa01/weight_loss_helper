```python
import numpy as np
import weight_loss_player
import matplotlib.pyplot as plt
from std_policies import eps_greeedy_policy
from mpl_toolkits.mplot3d import Axes3D
from tqdm import tqdm
%matplotlib inline
```

```python
player = weight_loss_player.WeighLossPlayer()
action_space_description,obesrvation_space_description = player.get_space_description()
n_action_states = len(action_space_description)
n_obs_states =len(obesrvation_space_description)
```

```python
print('Activity Levels: {0}'.format(player.activity_levels))
print('Calorie Intake Levels: {0}'.format(player.calorie_intake_levels))
print('Calorie Quality Levels: {0}'.format(player.calorie_quality_levels))
print('Action Space: {0}'.format(n_action_states))
print('Observation Space: {0}'.format(n_obs_states))
```

```python
def get_trends(act,n_days=100):
    weights=[]
    satiety=[]
    motivation=[]
    player.reset()
    print(act)
    for i in range(n_days):
        player.step(act,prob_of_accept_override=1)
        weights.append(player.weight_t)
        satiety.append(player.satiety_t)
        motivation.append(player.motivation_t)
    return np.array(weights),np.array(satiety),np.array(motivation)
```

```python
f,ax=plt.subplots(3,1)
for cal_qual in player.calorie_quality_levels:
    action=['sedentary','achieve',cal_qual]
    w,s,m=get_trends(action)
    ax[0].plot(w,label=cal_qual)
    ax[0].set_ylabel('Weight')
    ax[1].plot(s,label=cal_qual)
    ax[1].set_ylabel('Satiety')
    ax[2].plot(m,label=cal_qual)
    ax[2].set_ylabel('Motivation')
ax[1].legend()
f.suptitle("Model States Dependence on Calorie Quality ", fontsize=14)
plt.savefig('Calorie_quality')
plt.show()
```

```python
f,ax=plt.subplots(3,1)
for cal_level in player.calorie_intake_levels:
    action=['sedentary',cal_level,'normal']
    w,s,m=get_trends(action)

    ax[0].plot(w,label=cal_level)
    ax[0].set_ylabel('Weight')
    ax[1].plot(s,label=cal_level)
    ax[1].set_ylabel('Satiety')
    ax[2].plot(m,label=cal_level)
    ax[2].set_ylabel('Motivation')
    
ax[1].legend()
f.suptitle("Model States Dependence on Calorie Intake ", fontsize=14)
plt.savefig('Calorie_intake')
plt.show()
```

```python
f,ax=plt.subplots(3,1)
for act_level in player.activity_levels:
    action=[act_level,'achieve','normal']
    w,s,m=get_trends(action)
    ax[0].plot(w,label=act_level)
    ax[0].set_ylabel('Weight')
    ax[1].plot(s,label=act_level)
    ax[1].set_ylabel('Satiety')
    ax[2].plot(m,label=act_level)
    ax[2].set_ylabel('Motivation')
    
ax[1].legend()
f.suptitle("Model States Dependence on Activity", fontsize=14)
plt.savefig('Activity_level')
plt.show()
```
